# Problem Statement

We are providing you with a folder that has two subfolders,each having about 50 short clips of people jogging and walking respectively in varying environments
The task here is to load the videos,extract suitable features and train a Machine Learning model to classify each video as walking or jogging.


# Solution 

### Below code is to import the neccessary libraries 

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import glob, os
from sklearn.model_selection import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score
from skimage import feature
from sklearn.decomposition  import PCA
from sklearn.ensemble import RandomForestClassifier
```

## Reading Video data from files

### Reading walking videos from walking folder

####  Below code to read each data file from folder and converting to gray image. After that its finding histogram f Gradient for each image.

```python
os.chdir(r"C:\Users\Sarvani\myexp\data\signy\Walking-20180625T045208Z-001/Walking")
garyImagesWalking = []
for counter, file in enumerate(glob.glob("person*")):
   
    cap = cv2.VideoCapture(file)
    if(cap.isOpened() == False):
        print("Error opening video stream or file")
    ret, img = cap.read()
    gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    H = feature.hog(gray,orientations=9,pixels_per_cell=(10,10),cells_per_block=(2,2),transform_sqrt=True,block_norm='L2-Hys')
    garyImagesWalking.append(H)
    
    cap.release()
   
```

```python
np.array(garyImagesWalking).shape
```

```python
walkingdata = pd.DataFrame(np.array(garyImagesWalking).reshape(-1,np.array(garyImagesWalking).shape[1]))
```

```python
walkingdata['label'] = 0
```

```python
walkingdata.head(10)
```

### Reading walking videos from Jogging folder

####  Below code to read each data file from folder and converting to gray image. After that its finding histogram f Gradient for each image.

```python
os.chdir(r"C:/Users/Sarvani/myexp/data/signy/jogging-20180625T044814Z-001/jogging")
garyImagesJogging = []
for counter, file in enumerate(glob.glob("person*")):
   
    cap = cv2.VideoCapture(file)
    if(cap.isOpened() == False):
        print("Error opening video stream or file")
    ret, img = cap.read()
    gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    H = feature.hog(gray,orientations=9,pixels_per_cell=(10,10),cells_per_block=(2,2),transform_sqrt=True,block_norm='L2-Hys')
    garyImagesJogging.append(H)
```

```python
np.array(garyImagesJogging).shape
```

```python
joggingdata = pd.DataFrame(np.array(garyImagesJogging).reshape(-1,np.array(garyImagesJogging).shape[0]))
```

```python
joggingdata['label'] = 1
```

```python
joggingdata.head(10)
```

# creating datarame 

```python
data = walkingdata.append(joggingdata)
```

```python
data.shape
```

```python
data.head(10)
```

## Sampling Data

```python
feature_data = data.drop('label')
target_data = data['label']
```

```python
trainX, testX, trainY, testY = train_test_split(feature_data,target_data, random_state=5, test_size=0.25)
```

## Deriving Principle componets for image datauong randomized PCA

```python
n_components = 1000
pca = PCA(n_components=n_components, whiten=True,svd_solver='randomized').fit(trainX)
```

```python
X_train_pca = pca.transform(trainX)
X_test_pca = pca.transform(testX)
```

## Training model uisng RandomForestClassifier

```python
print("Fitting the classifier to the training set")
param_grid = {'n_estimators': [10, 15, 25, 50, 100], }
clf = GridSearchCV(RandomForestClassifier(oob_score=True), param_grid)
clf = clf.fit(X_train_pca, trainY)
print("Best estimator found by grid search:")
print(clf.best_estimator_)
```

## Testing Model Performance 

```python
print("Predicting activity names on the test set")
y_pred = clf.predict(X_test_pca)
print(classification_report(testY, y_pred, target_names=['0','1']))
print('Confusion  Matrix:\n ',confusion_matrix(testY, y_pred))
print('Accuracy Score:%',accuracy_score(testY, y_pred)*100)
```

#### Note : 0 represents 'walking' class and 1 represents 'jogging' class
