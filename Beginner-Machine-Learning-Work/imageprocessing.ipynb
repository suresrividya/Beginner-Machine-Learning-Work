```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
import time
```

```python
# COLOR Palette

def main():
     
    img1 = np.zeros((512,512,3),np.uint8)
    WINDOWNAME = 'OPEN CV BGR COLOR PALETTE'
    cv2.namedWindow(WINDOWNAME)
    
    cv2.createTrackbar("B",WINDOWNAME,0,255,cv2.setTrackbarPos)
    cv2.createTrackbar("G",WINDOWNAME,0,255,cv2.setTrackbarPos)
    cv2.createTrackbar("R",WINDOWNAME,0,255,cv2.setTrackbarPos)
    
    while(True):
        
        cv2.imshow(WINDOWNAME,img1)
        
        
        if cv2.waitKey(1) == 27:
            break
            
        B = cv2.getTrackbarPos("B",WINDOWNAME)
        G = cv2.getTrackbarPos("G",WINDOWNAME)
        R = cv2.getTrackbarPos("R",WINDOWNAME)
        
        img1[:] = [B,G,R]
        print(B,G,R)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()  
```

```python
img = np.zeros((512,512,3),np.uint8)
WINDOWNAME = "Drawing"
cv2.namedWindow(WINDOWNAME)

def draw(event,x,y,flags,param):
    
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img,(x,y),40,(255,126,56),-1)
        
    if event == cv2.EVENT_MBUTTONDOWN:
        cv2.circle(img,(x,y),20,(0,255,127),-1)
        
    
    
    
cv2.setMouseCallback(WINDOWNAME,draw)

def main():
    
    while(True):
        
        cv2.imshow(WINDOWNAME,img)
        
        
        if cv2.waitKey(1) == 27:
            break
            
    cv2.destroyAllWindows()  
    
if __name__ == "__main__":
    main() 
```

```python
 
def main():
    
    cap = cv2.VideoCapture(0)
    if cap.isOpened():
            ret, frame = cap.read()
            img = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            
            plt.imshow(img)
            plt.show()
    else:
            ret = False
        
    cap.release()
    
if __name__ == "__main__":
    main() 
```

```python
from skimage.transform import resize
from skimage.io import imread,imsave

def main():
  
    path = r"C:\Users\Sarvani\Picture\Dog\train\n_images"
    img1 = cv2.imread('data/vidya1.jpg',1)
    img2 = cv2.imread('data/vidya2.jpg',1)
    
   # img1 = cv2.resize(img1, (28,28),interpolation = )
    img1 = cv2.resize(img1, (256,256),interpolation= cv2.INTER_AREA)
    img2 = cv2.resize(img2, (256,256),interpolation= cv2.INTER_AREA)
    
    fps = 25
    height, width, layers = img1.shape
    size = (width,height)
    out = cv2.VideoWriter('datatransition.avi',cv2.VideoWriter_fourcc(*'DIVX'),fps,size)
    for i in np.linspace(0,1,100):
        
        alpha = i
        beta = 1-alpha
        output = cv2.addWeighted(img1,alpha,img2,beta,0)
        cv2.imshow('Transition', output)
        out.write(output)
        time.sleep(0.1)
        
        if cv2.waitKey(1) == 27:
            break
            
    cv2.destroyAllWindows()
    
    
if __name__ == "__main__":
    main()   
```
